```python
import pandas as pd 
```

```python
data_en = pd.read_csv("datasets_train/data_en.csv")
```

```python
data_en_WJ = data_en[data_en["safety"]==1]
data_en_wo_WJ = data_en[data_en["safety"]==0]
```

```python
N = len(data_en_WJ)
rs = 42
for n in range(10, 100, 10):
    print(n)
    print(N*(n/100))
    data_selected=data_en_WJ.groupby('category').apply(lambda x: x.sample(int(N*(n/100)/4), replace=False, random_state=rs))
    data = pd.concat([data_en_wo_WJ, data_selected])
    data.to_csv(f"datasets_train/data_en_{n}.csv")
```
